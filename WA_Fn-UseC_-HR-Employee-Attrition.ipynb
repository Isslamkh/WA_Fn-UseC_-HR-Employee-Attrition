```python
df['Gender'] = df['Gender'].replace({'Male': 0, 'Female': 0})
df['Attrition'] = df['Attrition'].replace({'Yes': 1, 'No': 0})
```

```python
df.columns
```

```python
prefix_col = ['Business', 'Depart', 'Education', 'JobRole', 'MaritalStatus', 'OverTime', 'Over18']
dummy_col = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus', 'OverTime', 'Over18']
df = pd.get_dummies(df, prefix=prefix_col, columns = dummy_col)
```

```python
test_data = df.tail(70)
```

```python
data = df.head(1400)
```

```python
data['Attrition'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
```

```python
X = data.drop(columns = ['Attrition'])
y = data['Attrition']
```

```python
X.shape
```

```python
y.shape
```

```python
X, y = sm.fit_resample(X, y)
```

```python
X.shape
```

```python
y.shape
```

```python
from xgboost import XGBClassifier
xg = XGBClassifier()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
xg.fit(X_train, y_train)
```

```python
y_pred = xg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score
accuracy_score = accuracy_score(y_test, y_pred)
recall_score = recall_score(y_test, y_pred)
f1_score = f1_score(y_test, y_pred)
```

```python
print('accuracy_score:', accuracy_score)
print('recall_score:', recall_score)
print('f1_score:', f1_score)
```
